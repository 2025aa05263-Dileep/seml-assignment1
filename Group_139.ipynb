## Problem
Predict if a student will **pass** using: study hours, attendance %, assignments done, previous score.

```python
import sys
from pathlib import Path
ROOT = Path.cwd()
sys.path.insert(0, str(ROOT))
```

```python
# Step 1: Generate sample data
from src.pipeline import StudentDataPipeline
import pandas as pd

DATA = ROOT / "data" / "students.csv"
StudentDataPipeline.generate_sample(DATA, n=500)
df = pd.read_csv(DATA)
print(f"Students: {len(df)} | Pass rate: {df['passed'].mean():.1%}")
df.head()
```

```python
# Step 2: Pipeline — clean & transform
pipe = StudentDataPipeline(DATA)
x_train, x_test, y_train, y_test = pipe.load()
print(f"Train: {len(x_train)} | Test: {len(x_test)} | Features: {pipe.FEATURES}")
```

```python
# Step 3: Train model
from src.train import train_model

MODEL = ROOT / "models" / "pass_model.joblib"
metrics = train_model(DATA, MODEL)
print("Metrics:", metrics)
```

```python
# Step 4: Predict (Microservice)
from src.predict import PassPredictor

pred = PassPredictor(MODEL)
print("Low-performing student:", pred.predict(5, 55, 3, 35))
print("High-performing student:", pred.predict(25, 90, 9, 75))
```

```python
# Step 5: Plot metrics (screenshot for report)
import matplotlib.pyplot as plt

plt.bar(metrics.keys(), metrics.values(), color="#1565c0")
plt.title("StudySmart Model Metrics (Group 139)")
plt.ylim(0, 1)
plt.show()
```
